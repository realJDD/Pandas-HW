```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_number_players = len(purchase_data['SN'].unique())
total_number_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
ave_price = purchase_data['Price'].mean()
unique_items = len(purchase_data['Item ID'].unique())
number_purchases = purchase_data['Purchase ID'].count()
total_revenue = purchase_data['Price'].sum()

columns=[
    "Number of Unique Items",
    "Average Price",
    "Number of Purchases",
    "Total Revenue"
]
summary_df = pd.DataFrame(index=[0], columns=columns)
summary_df['Number of Unique Items'] = unique_items
summary_df['Average Price'] = ave_price
summary_df['Average Price'] = summary_df['Average Price'].astype(float).map("${:.2f}".format)
summary_df['Number of Purchases'] = number_purchases
summary_df['Total Revenue'] = total_revenue
summary_df['Total Revenue'] = summary_df['Total Revenue'].astype(float).map("${:.2f}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
grouped_df = purchase_data.groupby(by=['SN'], as_index=False).first()

#Male data
male = grouped_df.loc[grouped_df['Gender'] == 'Male', :]
male_count = male['Gender'].count()

#Female data
female = grouped_df.loc[grouped_df['Gender'] == 'Female', :]
female_count = female['Gender'].count()

#Other data
other = grouped_df.loc[grouped_df['Gender'] == 'Other / Non-Disclosed', :]
other_count = other['Gender'].count()

#Total
total_count = male_count + female_count + other_count

#gender dataframe fill information
percentage_player = [(male_count/total_count)*100, (female_count/total_count)*100,(other_count/total_count)*100]
gdf_columns = ['Total Count', 'Percentage of Players']
gdf_index = ('Male', 'Female', 'Other / Non-Disclosed')
count = [male_count, female_count, other_count]

gender_df = pd.DataFrame(index = gdf_index, columns=gdf_columns)
gender_df['Total Count'] = count
gender_df['Percentage of Players'] = percentage_player
gender_df['Percentage of Players'] = gender_df['Percentage of Players'].astype(float).map("{:.2f}".format)
gender_df
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Dataframes
male_df = purchase_data.loc[purchase_data['Gender'] == 'Male', :]
female_df = purchase_data.loc[purchase_data['Gender'] == 'Female', :]
other_df = purchase_data.loc[purchase_data['Gender'] == 'Other / Non-Disclosed', :]

#Purchase Counts
pc_male = male_df['Purchase ID'].count()
pc_female = female_df['Purchase ID'].count()
pc_other = other_df['Purchase ID'].count()

#Avg Purchase Price
app_male = male_df['Price'].mean()
app_female = female_df['Price'].mean()
app_other = other_df['Price'].mean()

#Total Purchase Value
tpv_male = male_df['Price'].sum()
tpv_female = female_df['Price'].sum()
tpv_other = other_df['Price'].sum()

#Groupby to Gather Unique SN Count
group_male = male_df.groupby(by=['SN'], as_index=False).first()
group_male_count = group_male['Gender'].count()

group_female = female_df.groupby(by=['SN'], as_index=False).first()
group_female_count = group_female['Gender'].count()

group_other = other_df.groupby(by=['SN'], as_index=False).first()
group_other_count = group_other['Gender'].count()

#Avg Total Purchase Per Person
atppp_male = tpv_male / group_male_count
atppp_female = tpv_female / group_female_count
atppp_other = tpv_other / group_other_count

#Purchasing Analysis Dataframe Fill Information
padf_columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']
padf_index = ('Female', 'Male', 'Other / Non-Disclosed')

#Build Purchasing Analysis Dataframe 
pa_df = pd.DataFrame(index=padf_index, columns=padf_columns)
pa_df['Purchase Count'] = pc_female, pc_male, pc_other
pa_df['Average Purchase Price'] = app_female, app_male, app_other
pa_df['Average Purchase Price'] = pa_df['Average Purchase Price'].astype(float).map("${:.2f}".format)
pa_df['Total Purchase Value'] = tpv_female, tpv_male, tpv_other
pa_df['Total Purchase Value'] = pa_df['Total Purchase Value'].astype(float).map("${:.2f}".format)
pa_df['Avg Total Purchase per Person'] = atppp_female, atppp_male, atppp_other
pa_df['Avg Total Purchase per Person'] = pa_df['Avg Total Purchase per Person'].astype(float).map("${:.2f}".format)
pa_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#group DF by SN
grouped_ad_df = purchase_data.groupby(by=['SN'], as_index=False).first()

#Establish bins and labels
bins = [0, 9, 14, 19, 24, 29, 34, 39, 199]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]

#Bin the Data & find the value counts
binned = pd.cut(grouped_ad_df['Age'], bins, labels=labels)
binned_count = pd.value_counts(binned, sort=False)

#Add Binned Data to  New DataFrame & Rename Column
ad_df = pd.DataFrame(binned_count)
ad_df = ad_df.rename(columns={"Age":"Total Count"})

#Grab the Sum of the 'Total Count' Column
sum = ad_df['Total Count'].sum()

#Build Dataframe
ad_df['Percentage of Players'] = (ad_df['Total Count'] / sum)*100
ad_df['Percentage of Players'] = ad_df['Percentage of Players'].astype(float).map("{:.2f}".format)
ad_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Establish bins and labels
bins = [0, 9, 14, 19, 24, 29, 34, 39, 199]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#bin the data & find the value counts
binned_df = purchase_data
binned_df['Bins'] = pd.cut(purchase_data['Age'], bins, labels=labels)
binned_df = binned_df.groupby('Bins')

#Find the Purchase Count
pa_p_count = binned_df['SN'].count()

#Find the Total Purchase Value
pa_sum = binned_df['Price'].sum()

#Summary
pa_summary = pd.DataFrame({'Purchase Count': pa_p_count,
                           'Average Purchase Price': pa_sum / pa_p_count,
                           'Total Purchase Value': pa_sum,
                           'Avg Total Purchase per Person': pa_sum / binned_count
                          })

#Format
pa_summary['Total Purchase Value'] = pa_summary['Total Purchase Value'].astype(float).map("${:,.2f}".format)
pa_summary['Average Purchase Price'] = pa_summary['Average Purchase Price'].astype(float).map("${:.2f}".format)
pa_summary['Avg Total Purchase per Person'] = pa_summary['Avg Total Purchase per Person'].astype(float).map("${:.2f}".format)

avg
pa_summary.head()

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
mpi_df = purchase_data
mpi_df = mpi_df.loc[:,['Item ID','Item Name','Price']]

#perform groupby operation
mpi_df = mpi_df.groupby(['Item ID', 'Item Name'])

#find sum and purchase count
mpi_sum = mpi_df['Price'].sum()
mpi_p_count = mpi_df['Item Name'].count()
mpi_i_price = mpi_df['Price'].mean()

mpi_summary = pd.DataFrame({'Purchase Count': mpi_p_count,
                            'Item Price': mpi_i_price,
                            'Total Purchase Value':mpi_sum
                            })

#Formatting
mpi_summary['Item Price'] = mpi_summary['Item Price'].astype(float).map("${:,.2f}".format)
mpi_summary['Total Purchase Value'] = mpi_summary['Total Purchase Value'].astype(float).map("${:,.2f}".format)
mpi_summary = mpi_summary.sort_values(by=['Purchase Count'], ascending=False)
mpi_summary.head()
```
